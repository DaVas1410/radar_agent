```python
# SYSTEMATIC TESTING: Multiple Technology Radar Topics - SIMPLE VERSION
# No pandas dependency, using native Python data structures

import sys
import json
import os
import time
from datetime import datetime

# Ensure clean module loading
for module in list(sys.modules.keys()):
    if module.startswith('agent'):
        del sys.modules[module]

from agent.graph import graph as radar_agent
from langchain_core.messages import HumanMessage

print("SYSTEMATIC RADAR TESTING - MULTIPLE TOPICS")
print("=" * 60)
print("Testing 7 different technology domains with standardized metrics")
print()

# Define test topics with business relevance
test_topics = [
    "DevOps Tools 2025",
    "Frontend Frameworks", 
    "Database Technologies",
    "AI/ML Platforms",
    "Cybersecurity Tools",
    "Mobile Development",
    "Cloud Services"
]

# Test configuration
target_elements = 12  # Small number for fast testing
max_loops = 1         # Single iteration for speed

# Results storage
test_results = []
all_radar_data = {}

print(f"Configuration: {target_elements} elements per radar, {max_loops} research loop")
print()

# Execute tests for each topic
for i, topic in enumerate(test_topics, 1):
    print(f"TEST {i}/7: {topic}")
    print("-" * 40)
    
    # Initialize test state
    test_state = {
        "messages": [HumanMessage(content=topic)],
        "max_research_loops": max_loops,
        "initial_search_query_count": 2,
        "radar_topic": topic,
        "target_element_count": target_elements,
        "radar_elements": []
    }
    
    # Track execution time
    start_time = time.time()
    
    try:
        # Execute radar construction
        result = radar_agent.invoke(test_state)
        
        # Calculate execution metrics
        execution_time = time.time() - start_time
        
        # Extract results
        elements_found = len(result.get('radar_elements', []))
        sources_count = len(result.get('sources_gathered', []))
        radar_json = result.get('radar_json', {})
        
        # Get summary from messages
        summary_text = ""
        if result.get('messages') and len(result['messages']) > 0:
            summary_text = result['messages'][-1].content
        
        # Calculate token estimates (approximate)
        summary_tokens = len(summary_text.split()) * 1.3 if summary_text else 0
        json_size = len(json.dumps(radar_json)) if radar_json else 0
        
        # Extract statistics safely
        quadrant_stats = {}
        ring_stats = {}
        
        if radar_json and 'statistics' in radar_json:
            quadrant_stats = radar_json['statistics'].get('quadrants', {})
            ring_stats = radar_json['statistics'].get('rings', {})
        
        # Create result dictionary
        test_result = {
            'topic': topic,
            'time_seconds': round(execution_time, 2),
            'elements_found': elements_found,
            'target_elements': target_elements,
            'accuracy_percent': round((elements_found / target_elements * 100), 1),
            'sources_analyzed': sources_count,
            'summary_length': len(summary_text),
            'token_estimate': int(summary_tokens),
            'json_size_bytes': json_size,
            'techniques': quadrant_stats.get('Techniques', 0),
            'tools': quadrant_stats.get('Tools', 0),
            'platforms': quadrant_stats.get('Platforms', 0),
            'frameworks': quadrant_stats.get('Languages & Frameworks', 0),
            'adopt': ring_stats.get('Adopt', 0),
            'trial': ring_stats.get('Trial', 0),
            'assess': ring_stats.get('Assess', 0),
            'hold': ring_stats.get('Hold', 0)
        }
        
        test_results.append(test_result)
        
        # Store complete radar data for plotting
        all_radar_data[topic] = {
            'radar_json': radar_json,
            'summary': summary_text,
            'elements': result.get('radar_elements', [])
        }
        
        print(f"SUCCESS: {elements_found} elements in {execution_time:.1f}s")
        print(f"Accuracy: {test_result['accuracy_percent']}% of target")
        
    except Exception as e:
        print(f"ERROR: {e}")
        # Store error result
        test_result = {
            'topic': topic,
            'time_seconds': round(time.time() - start_time, 2),
            'elements_found': 0,
            'target_elements': target_elements,
            'accuracy_percent': 0,
            'sources_analyzed': 0,
            'summary_length': 0,
            'token_estimate': 0,
            'json_size_bytes': 0,
            'error': str(e)[:100]
        }
        test_results.append(test_result)
    
    print()

print("COMPREHENSIVE TEST RESULTS")
print("=" * 60)
print()

# Create manual table display
print("PERFORMANCE SUMMARY:")
print(f"{'Topic':<20} {'Time(s)':<8} {'Elements':<9} {'Accuracy%':<10} {'Tokens':<8} {'Sources':<8}")
print("-" * 70)
for result in test_results:
    if 'error' not in result:
        print(f"{result['topic']:<20} {result['time_seconds']:<8} {result['elements_found']:<9} {result['accuracy_percent']:<10} {result['token_estimate']:<8} {result['sources_analyzed']:<8}")
    else:
        print(f"{result['topic']:<20} ERROR")

print()
print("QUADRANT DISTRIBUTION:")
print(f"{'Topic':<20} {'Tech':<6} {'Tools':<6} {'Plat':<6} {'Frame':<6}")
print("-" * 50)
for result in test_results:
    if 'error' not in result:
        print(f"{result['topic']:<20} {result['techniques']:<6} {result['tools']:<6} {result['platforms']:<6} {result['frameworks']:<6}")

print()
print("ADOPTION RISK DISTRIBUTION:")
print(f"{'Topic':<20} {'Adopt':<6} {'Trial':<6} {'Assess':<6} {'Hold':<6}")
print("-" * 50)
for result in test_results:
    if 'error' not in result:
        print(f"{result['topic']:<20} {result['adopt']:<6} {result['trial']:<6} {result['assess']:<6} {result['hold']:<6}")

# Calculate aggregate statistics
successful_tests = [r for r in test_results if 'error' not in r]

if successful_tests:
    avg_time = sum(r['time_seconds'] for r in successful_tests) / len(successful_tests)
    avg_accuracy = sum(r['accuracy_percent'] for r in successful_tests) / len(successful_tests)
    total_tokens = sum(r['token_estimate'] for r in successful_tests)
    total_elements = sum(r['elements_found'] for r in successful_tests)
    total_time = sum(r['time_seconds'] for r in successful_tests)
    
    print()
    print("AGGREGATE METRICS:")
    print(f"Successful tests: {len(successful_tests)}/{len(test_results)}")
    print(f"Average execution time: {avg_time:.1f} seconds")
    print(f"Average accuracy: {avg_accuracy:.1f}%")
    print(f"Total tokens estimated: {total_tokens:,.0f}")
    print(f"Total elements analyzed: {total_elements}")
    print(f"Processing rate: {total_elements/total_time:.1f} elements/second")

print()
print("Data ready for visualization:")
print("- test_results: List of dictionaries with all metrics")
print("- all_radar_data: Complete radar JSON data for each topic")
print(f"- {len(successful_tests)} successful radar analyses completed")

# Show first result as example
if test_results:
    print()
    print("Sample result structure:")
    sample = test_results[0]
    for key, value in sample.items():
        if key != 'error':  # Skip error field if present
            print(f"  {key}: {value}")

print()
print("Ready for plotting and analysis!")
```
